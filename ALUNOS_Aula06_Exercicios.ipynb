# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
import matplotlib.pyplot as plt
```

```python
meses = [f'Mês {i+1}' for i in range(12)]
precos = np.random.uniform(10, 100, size=12).round(2)
```

```python
df_precos = pd.DataFrame({
    'Mês': meses,
    'Preço': precos
})
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_precos['Mês'], df_precos['Preço'], marker='o', linestyle='--', color='b')
plt.title('Variação de preços de um produto durante 12 Meses')
plt.xlabel('Mês')
plt.ylabel('Preço')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
num_pessoas = 100
alturas2 = np.random.uniform(150, 190, size=num_pessoas)
pesos2 = np.random.uniform(50, 100, size=num_pessoas)

df_altpeso = pd.DataFrame({
    'Altura': alturas2,
    'Peso': pesos2
})

plt.figure(figsize=(10, 6))
sns.regplot(x='Altura', y='Peso', data=df_altpeso, scatter_kws={'color': 'skyblue', 's': 50}, line_kws={'color': 'purple', 'linewidth': 2})

plt.title('Relação entre Altura e Peso + Linha de Regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
data = {
    'V1': np.random.rand(50) * 100,
    'V2': np.random.rand(50) * 100,
    'V3': np.random.rand(50) * 100,
    'V4': np.random.rand(50) * 100
}
```

```python
df_heatmap = pd.DataFrame(data)

coorelacao = df_heatmap.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(coorelacao, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)

plt.title('Heatmap de Correlação entre diferentes Variáveis')
plt.show()
```

# 20 - Pairplot

Crie um pairplot para explorar relações entre várias variáveis em um conjunto de dados fictício.

```python
data = {
    'VA': np.random.rand(100) * 50,
    'VB': np.random.rand(100) * 50,
    'VC': np.random.rand(100) * 50,
    'VD': np.random.rand(100) * 50
}

df_pairplot = pd.DataFrame(data)
```
